```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

# 데이터 로드 및 전처리
KS = pd.read_csv('KS_DSPSN_SVCH_UTILIIZA_CRSTAT_INFO_202410.csv')
KS2 = pd.read_csv('KS_SVCH_UTILIIZA_CRSTAT_INFO_202410.csv')

KS = KS[['FCLTY_NM', 'ITEM_NM','TROBL_TY_NM','COURSE_PRC','COURSE_NM','COURSE_NO','CTPRVN_NM','SIGNGU_NM']]
KS.fillna("0", inplace=True)
KS['TROBL_TY_NM'] = KS['TROBL_TY_NM'].apply(lambda x: 0 if x == "0" else 1)

KS2 = KS2[['FCLTY_NM', 'ITEM_NM','COURSE_PRC','COURSE_NM','COURSE_NO','CTPRVN_NM','SIGNGU_NM']]
KS_sum = pd.concat([KS, KS2])
KS_sum.drop_duplicates(subset="COURSE_NO", keep='first', inplace=True)

# 인코더 설정
item_nm = LabelEncoder()
trobl_ty = LabelEncoder()
CTPRVN_NM = LabelEncoder()

X = KS_sum[['ITEM_NM','TROBL_TY_NM','CTPRVN_NM']]
y = KS_sum[['COURSE_NO']]

# 레이블 인코딩
X['CTPRVN_NM'] = CTPRVN_NM.fit_transform(X['CTPRVN_NM'])
X['TROBL_TY_NM'] = trobl_ty.fit_transform(X['TROBL_TY_NM'])
X['ITEM_NM'] = item_nm.fit_transform(X['ITEM_NM'])

# 모델 훈련
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# 헬스 항목 정의
헬스_항목들 = ['헬스']  # 여러 헬스 항목
결과 = []

# 여러 헬스 항목에 대해 예측 수행
for 항목 in 헬스_항목들:
    스포츠 = item_nm.transform([항목])[0]
    장애 = trobl_ty.transform(['1'])[0]
    시도명 = CTPRVN_NM.transform(['서울'])[0]
    input_data = [[스포츠, 장애, 시도명]]

    # 예측 수행
    예측값 = model.predict(input_data)[0]

    # 예측값으로 KS_sum에서 필터링
    필터링된_데이터 = KS_sum[KS_sum['COURSE_NO'] == 예측값]
    
    # 결과를 추가
    if not 필터링된_데이터.empty:
        결과.append({
            '항목': 항목,
            '예측': 예측값,
            '필터링된 데이터': 필터링된_데이터.to_dict(orient='records')  # DataFrame을 딕셔너리로 변환
        })

# 결과 출력
for r in 결과:
    print(f"항목: {r['항목']}, 예측: {r['예측']}, 필터링된 데이터: {r['필터링된 데이터']}")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

# 이미 전처리된 데이터 로드
KS_sum = pd.read_csv('KS_sum.csv')

# 레이블 인코더 설정
item_nm = LabelEncoder()
trobl_ty = LabelEncoder()
CTPRVN_NM = LabelEncoder()

# X, y 설정
X = KS_sum[['ITEM_NM', 'TROBL_TY_NM', 'CTPRVN_NM']]
y = KS_sum['COURSE_NO']

# 레이블 인코딩
X['ITEM_NM'] = item_nm.fit_transform(X['ITEM_NM'])
X['TROBL_TY_NM'] = trobl_ty.fit_transform(X['TROBL_TY_NM'])
X['CTPRVN_NM'] = CTPRVN_NM.fit_transform(X['CTPRVN_NM'])

# 모델 훈련
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# 헬스 항목 리스트
헬스_항목들 = ['헬스', '요가', '필라테스']  # 추가적인 데이터

# 결과 저장 리스트
결과 = []

# 여러 항목에 대해 예측
for 항목 in 헬스_항목들:
    try:
        스포츠 = item_nm.transform([항목])[0]  # 항목 인코딩
        장애 = trobl_ty.transform(['1'])[0]  # TROBL_TY_NM 값 (예: 장애 있음)
        시도명 = CTPRVN_NM.transform(['서울'])[0]  # 예: 서울

        # 입력 데이터 구성
        input_data = [[스포츠, 장애, 시도명]]
        예측값 = model.predict(input_data)[0]

        # 예측값 필터링
        필터링된_데이터 = KS_sum[KS_sum['COURSE_NO'] == 예측값]

        if not 필터링된_데이터.empty:
            결과.append({
                '항목': 항목,
                '예측 COURSE_NO': 예측값,
                '필터링된 데이터': 필터링된_데이터[['FCLTY_NM', 'COURSE_NM', 'CTPRVN_NM', 'SIGNGU_NM']].to_dict(orient='records')
            })
    except Exception as e:
        print(f"항목 {항목} 처리 중 오류 발생: {e}")

# 결과 저장 및 출력
결과_df = pd.DataFrame(결과)
결과_df.to_csv('결과.csv', index=False)
print(결과_df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import random

# 데이터 로드
KS_sum = pd.read_csv('KS_sum.csv')

# 레이블 인코더 설정
item_nm = LabelEncoder()
trobl_ty = LabelEncoder()
CTPRVN_NM = LabelEncoder()

# 레이블 인코딩
KS_sum['ITEM_NM'] = item_nm.fit_transform(KS_sum['ITEM_NM'])
KS_sum['TROBL_TY_NM'] = trobl_ty.fit_transform(KS_sum['TROBL_TY_NM'])
KS_sum['CTPRVN_NM'] = CTPRVN_NM.fit_transform(KS_sum['CTPRVN_NM'])

# 항목 리스트
항목들 = ['축구(풋살)', '탁구', '수영']  # 원하는 항목 추가 가능

# 결과 저장 리스트
결과 = []

# 조건에 따라 무작위 추천
for 항목 in 항목들:
    try:
        # 조건에 맞는 데이터 필터링
        필터링된_데이터 = KS_sum[
            (KS_sum['ITEM_NM'] == item_nm.transform([항목])[0]) & 
            (KS_sum['TROBL_TY_NM'] == 0) &  # TROBL_TY_NM == 1 조건
            (KS_sum['CTPRVN_NM'] == CTPRVN_NM.transform(['대구'])[0])  # 서울 조건
        ]

        # 데이터를 섞어서 무작위 추천
        섞인_데이터 = 필터링된_데이터.sample(frac=1, random_state=random.randint(0, 1000))  # 데이터 섞기
        추천_데이터 = 섞인_데이터.head(3)  # 상위 3개를 추천 (조정 가능)

        # 결과 저장
        결과.append({
            '항목': 항목,
            '추천 데이터': 추천_데이터[['FCLTY_NM', 'COURSE_NM', 'CTPRVN_NM', 'SIGNGU_NM']].to_dict(orient='records')
        })

    except Exception as e:
        print(f"항목 {항목} 처리 중 오류 발생: {e}")

# 결과 저장 및 출력

결과_df = pd.DataFrame(결과)
결과_df
```
